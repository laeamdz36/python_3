```python
"""Practica de creacion y lectura de archivos"""
import datetime as dt
import time as tm
import fileinput as fi
import os
import re 

# Obteniendo referencias a archivos
# built in funtion open() create a files object for read or write data to and from
# required the name of file and the mode ()read or write)
# syntax file_objetc = open(file_name, access_mode, buffering)
```

```python
# reading a file
file_name = "./logs.txt"
file_logs = open(file_name)
file_lines = file_logs.readlines()
for line in file_lines:
    print(line,end="")
file_logs.close()
```

```python
# creating a new file
print("writing a file")
file_new = open("./new_text_file.txt","w+")
file_new.write("Hello world\n")
file_new.write("Writing into file\n")
file_new.write("Yes dude!\n")

file_new.close()
print("Finishing")
```

```python
# updating file
file_path = "./new_text_file.txt"
update_text = open(file_path,"a")
new_text = "Hello world again"
update_text.write(new_text.capitalize())
update_text.close()
```

```python
# perform a loop and write into the file in append mode.

file_insert = open(file_path,"w")

for i,num in enumerate(range(1000)):
    tm.sleep(0.1)
    tstamp = dt.datetime.timestamp(dt.datetime.today())
    file_insert.write(f"{i} -> {num} at {str(tstamp)} _ result\n")
file_insert.close()
```

```python
# Context manager protocol

with open(file_path) as read_file:
    read_file_lines = read_file.readlines()
    for line in read_file_lines:
        print(line, end="")
```

```python
# files names
dc5_path = "./Data concentrator 5.txt"
dcs_path = "./Data concentrators.txt"

with fi.input(files = (dc5_path,dcs_path)) as f:
    line = f.readline()
    print(f"f.filename(): {f.filename()}")
    print(f"f.isfirstline: {f.isfirstline()}")
    print(f"f.lineno(): {f.lineno()}")
    print(f"f.filelineno(): {f.filelineno()}")
```
