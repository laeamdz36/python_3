```python
import pandas as pd
import numpy as np
import math as mt
import matplotlib.pyplot as plt

a = np.arange(0, 100)
df_1 = pd.Series(a*a*a)
# filtering values
graf=df_1[(df_1 < 4000) & (df_1 > 10)]
print(graf)
graf.plot(title="Impresion de valores filtrados", grid=1)
plt.show()
print(df_1)
```

```python
b=np.arange(1,100)
df_2=pd.Series(np.log(b))
series_2=pd.Series(b*b)
print(series_2)
df_2.plot(title="Log grafica",grid=1)
plt.show()
```

```python
sin_array=np.linspace(0,10,1000)
sin_serie=pd.Series(np.sin(sin_array))
sin_serie.plot(title="Grafica de seno",grid=1)
plt.show()
```
