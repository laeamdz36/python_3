# Operaciones aritmeticas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# plt.style.use('_mpl-gallery')

a= np.arange(0,250)
array_sin=a*np.sin(a)
fig=plt.plot(array_sin,)
plt.show()

```

```python
x_1 = np.linspace(0, 10, 100)
y_1 = 4 + 2 * np.sin(2*x_1)
fig, ax = plt.subplots()
ax.plot(x_1, y_1, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```
