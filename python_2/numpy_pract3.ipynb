```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

a=np.arange(0,1000)
a_sqrt=np.sqrt(a)
array_sqrt=10*a_sqrt*np.sin(a_sqrt)

fig=plt.plot(array_sqrt)
plt.grid(1)
plt.title("FUncion exponencial seno")
plt.show()
```

```python
b = np.arange(0, 1000)
b_sqrt = np.sqrt(b)
array_sqrt_b = 10*b_sqrt

fig = plt.plot(array_sqrt_b)
plt.grid(1)
plt.title("FUncion exponencial seno")
plt.show()
```

# Increment and defcrement operands

# Universal functions

```python
array_5 = np.arange(100)
log_array = np.log(array_5)
fig_2 = plt.plot(log_array)
plt.show()
```

# Aggregate functions

```python
array_6 = np.arange(20)
print(array_6.sum())
print(array_6.max())
print(array_6.min())
print(array_6.mean())
array_6.std()
```

# Indexing, slicing and itering

```python
array_26=np.arange(10,16)
print(array_26)
array_26[4]
print(array_26[[1,3]])
```

# Slicing

```python
array_26[1:3]
array_27=np.arange(0,21)

print(array_27[7:15:3])
print(array_27[::5])
```

```python
# impresion de valores dentro del array
print(array_27)
# Impresion de los ultimos dos elementos dentro del array
print(array_27[-2:])
# impresion de elementos del array en saltos de 3 en 3
print(array_27[0:array_27.size:3])
```

```python
# impresion del array de 2 en 2
print(array_27[::2])
```

```python
# Crear array de 30 elementos y reformarlo en una matriz de 6 x 5
array_30=np.arange(0,30).reshape((6,5))
print(array_30)
```

```python
# Slicing matriz
print("Impresion de la ultima columna")
print(array_30[:,-1:])
```

```python
print("Impresion de la matriz")
print(array_30)
print("Impresion de la penultima columna")
print(array_30[:,-2:-1])
print("Impresion de las primeras dos filas")
print(array_30[:2,:])
print("impresion de las ultimas dos filas")
print(array_30[-2:,:])
```

```python
# Si las filas o columnas no son continuas se puede especificar el index
print(array_30[[2,4],-2:])
```

```python
# Uso de la funcion apply along axis par afunciones aggregate
print(array_30)
print("\nImpresion de suma por columna\n")
print(np.apply_along_axis(np.sum, axis=0, arr=array_30))
print("\nImpresion de suma por fila\n")
print(np.apply_along_axis(np.sum, axis=1, arr=array_30))

print("\nImpresion de promedio por columna\n")
print(np.apply_along_axis(np.mean, axis=0, arr=array_30))
```

```python
print(np.apply_along_axis(lambda x: x * 10, axis=0, arr=array_30))
```

```python
print(np.apply_along_axis(lambda x: x * 10, axis=1, arr=array_30))
```

```python
print(np.apply_along_axis(lambda x : x /10,axis=0,arr=array_30))
```
