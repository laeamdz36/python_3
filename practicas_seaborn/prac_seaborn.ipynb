```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg").dropna()
cars.shape
```

```python
cars.head()
```

```python
plt.style.use("ggplot")
sns.kdeplot(cars.horsepower,fill=True)
```

```python
plt.style.use("seaborn-v0_8-darkgrid")
sns.kdeplot(data=cars,x=cars.horsepower,y=cars.mpg,color="b");
```

```python
plt.style.use("seaborn-v0_8-darkgrid")
sns.kdeplot(data=cars,x=cars.horsepower,y=cars.mpg,color="m",n_levels=15);
```

```python
plt.style.use("seaborn-v0_8-darkgrid")
sns.kdeplot(data=cars,x=cars.horsepower,y=cars.mpg,color="b",n_levels=10,fill=True);
```

```python
#plt.figure(figsize=(10,10),dpi=200)
plt.style.use("seaborn-v0_8-darkgrid")
sns.kdeplot(data=cars,x=cars.horsepower,y=cars.mpg,color="b",n_levels=10,fill=True,thresh=0.01,cbar=True)
plt.show()
```
